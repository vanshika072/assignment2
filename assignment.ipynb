# Question 2

```python
url='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'
```

```python
import requests
html_data=requests.get(url).text
html_data
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html_data,'html5lib')
soup
```

```python
tesla_revenue=pd.DataFrame(columns=['Date','Revenue'])
tesla_revenue
```

```python
for row in soup.find('tbody').find_all('tr'):
    col=row.find_all('td')
    date=col[0].text
    revenue=col[1].text
    tesla_revenue=tesla_revenue.append({'Date':date,'Revenue':revenue},ignore_index=True)
tesla_revenue    
```

```python
tesla_revenue['Revenue']=tesla_revenue['Revenue'].str.replace(r'\$|,', '')
tesla_revenue['Revenue']
```

```python
tesla_revenue.dropna(inplace=True)

tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
tesla_revenue
```

```python
tesla_revenue.tail(5)
```

```python
def make_graph(tesla_data, tesla_revenue, title):
    # Assuming you have the necessary code to plot graphs here
    # You can use libraries like matplotlib to create graphs
    
    # Example code for plotting Tesla stock data and revenue data
    import matplotlib.pyplot as plt
    
    # Assuming stock_data and revenue_data are properly formatted dataframes or arrays
    # Plot Tesla stock data
    plt.figure(figsize=(10, 6))
    plt.plot(tesla_data['Date'], tesla_data['Close'], label='Stock Close Price')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f'Tesla Stock Data - {title}')
    plt.legend()
    plt.grid(True)
    
    # Plot Tesla revenue data
    plt.figure(figsize=(10, 6))
    plt.plot(tesla_revenue['Date'], tesla_revenue['Revenue'], label='Revenue')
    plt.xlabel('Date')
    plt.ylabel('Revenue')
    plt.title(f'Tesla Revenue Data - {title}')
    plt.legend()
    plt.grid(True)
    
    plt.show()

# Call the make_graph function
make_graph(tesla_data, tesla_revenue, 'Tesla')
    
```
